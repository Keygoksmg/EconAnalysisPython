```python
import pandas as pd
import statsmodels.api as sm
```

```python
# import excel using Data1.xlsx, clear first sheet(夫婦就業) 
df = pd.read_excel('data/1_Data.xlsx', index_col=0)
df.head()
```

```python
# sum t年夫の育休 t1年妻の就業継続 t年妻の収入 夫就業理解
df.describe()
```

```python
# tabstat t1年妻の就業継続 , by(t年夫の育休)
# reg t1年妻の就業継続 t年夫の育休, r
# 表３
X = sm.add_constant(df['t年夫の育休'])
Y = df['t+1年妻の就業継続']
model_sm3 = sm.OLS(Y, X).fit()

# 結果を出力
model_sm3.summary()
```

```python
# reg t1年妻の就業継続 t年夫の育休 t年妻の収入, r
# 表４
X = sm.add_constant(df[['t年夫の育休', 't年妻の収入']])
Y = df['t+1年妻の就業継続']
model_sm4 = sm.OLS(Y, X).fit()

# 結果を出力
model_sm4.summary()
```

```python
# reg t1年妻の就業継続 t年夫の育休 t年妻の収入 夫就業理解, r
# 表５
X = sm.add_constant(df[['t年夫の育休', 't年妻の収入', '夫就業理解']])
Y = df['t+1年妻の就業継続']
model_sm5 = sm.OLS(Y, X).fit()

# 結果を出力
model_sm5.summary()
```
