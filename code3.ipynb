```python
import pandas as pd
import statsmodels.api as sm
```

```python
# import excel using 3_Data.xlsx, clear first sheet(Sheet1) 
# 表１
df1 = pd.read_excel('data/3_Data.xlsx', index_col=0, sheet_name='Sheet1')
df1
```

```python
# tabstat wage, by(educ)
df1.groupby('educ').mean()
```

```python
# tabstat wage, by(female)
df1.groupby('female').mean()
```

```python
# reg wage female, r
# 表２
X = sm.add_constant(df1['female'])
Y = df1['wage']
model_sm2 = sm.OLS(Y, X).fit()

# 結果を出力
model_sm2.summary()
```

```python
# reg wage educ, r
# 表３
X = sm.add_constant(df1['educ'])
Y = df1['wage']
model_sm3 = sm.OLS(Y, X).fit()

# 結果を出力
model_sm3.summary()
```

```python
# reg wage female educ, r
# 表４
X = sm.add_constant(df1[['educ', 'female']])
Y = df1['wage']
model_sm4 = sm.OLS(Y, X).fit()

# 結果を出力
model_sm4.summary()
```

```python
# import excel using 3_Data.xlsx, clear first sheet(Sheet2) 
# 表５
df2 = pd.read_excel('data/3_Data.xlsx', index_col=0, sheet_name='Sheet2')
df2
```

```python
# reg wage female, r
# 表６
X = sm.add_constant(df2['female'])
Y = df2['wage']
model_sm6 = sm.OLS(Y, X).fit()

# 結果を出力
model_sm6.summary()
```

```python
# reg wage educ, r
# 表７
X = sm.add_constant(df2['educ'])
Y = df2['wage']
model_sm7 = sm.OLS(Y, X).fit()

# 結果を出力
model_sm7.summary()
```

```python
# reg wage educ female, r
# 表８
X = sm.add_constant(df2[['female', 'educ']])
Y = df2['wage']
model_sm8 = sm.OLS(Y, X).fit()

# 結果を出力
model_sm8.summary()
```

```python
# reg educ female, r
# 表９
X = sm.add_constant(df2['female'])
Y = df2['educ']
model_sm9 = sm.OLS(Y, X).fit()

# 結果を出力
model_sm9.summary()
```

```python
# predict residual, resid
residual = df2['educ'] - model_sm9.predict()
# reg wage residual, r
# 表１０
X = sm.add_constant(residual)
Y = df2['wage']
model_sm10 = sm.OLS(Y, X).fit()

# 結果を出力
model_sm10.summary()
```

```python
# reg wage educ if female==0, r
# reg wage educ if female==1, r
df_male = df2[df2['female'] == 0]
# 表１１、１２
X = sm.add_constant(df_male['educ'])
Y = df_male['wage']
model_sm11 = sm.OLS(Y, X).fit()

# 結果を出力
model_sm11.summary()
```

```python
# gen educ_female=educ*female
df2['educ_female'] = df2['educ'] * df2['female']

# reg wage educ_female educ female, r
# reg wage female##educ, r
# 表１３
X = sm.add_constant(df2[['educ_female', 'educ', 'female']])
Y = df2['wage']
model_sm13 = sm.OLS(Y, X).fit()

# 結果を出力
model_sm13.summary()
```

```python
# reg wage educ_female educ, r
# 表１５
X = sm.add_constant(df2[['educ_female', 'educ']])
Y = df2['wage']
model_sm15 = sm.OLS(Y, X).fit()

# 結果を出力
model_sm15.summary()
```
